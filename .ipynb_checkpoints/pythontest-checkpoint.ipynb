```python
import numpy as np
import pandas as pd
import pendulum
from datetime import datetime
from pathlib import Path
```

```python
start = pendulum.date(2017, 9, 1)
end = pendulum.date(2019, 1, 1)
dates_range = pendulum.period(start, end)
dfs = []
for date in dates_range.range("days"):
    path = "/home/huxley/data/s1-data/F01451_StoreSale_"
    path += date.format("%Y_%m_%d")
    file = Path(path)
    if file.is_file():
        df = pd.read_csv(path,
                         sep=";",
                         header=None,
                         usecols=[1, 2, 4, 5, 8, 9, 11, 13, 17, 18, 21, 22],
                         names=["date", "ean", "id", "name",
                                "category", "qty", "sales",
                                "profit", "unit", "vat",
                                "weight", "category-id"],
                         parse_dates=[0],
                         date_parser=lambda x: datetime.strptime(x, "%Y-%m-%d"),
                         encoding="windows-1250")
        dfs.append(df)
df = pd.concat(dfs)
df["month"] = df["date"].apply(lambda x: x.replace(day=1))
df.head()
```

```python
tmp_df = df[df["date"] >= datetime(2017, 10, 1)]
tmp_df["margin"] = tmp_df["profit"] / tmp_df["sales"]
tmp_df.head()
```
